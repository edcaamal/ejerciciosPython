```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
class reglasPPT:
  def __init__(self, piedra, papel, tijera):
    self.piedra = piedra
    self.papel = papel
    self.tijera = tijera

  def __str__(self):
    return f"{self.piedra}.{self.papel}.{self.tijera}"
  
  def regresaLista(self):
    return self.piedra, self.papel, self.tijera

personajes = ["piedra", "papel", "tijera"]
resultado = ["empate", "gana", "pierde"]
jugadas = []
jugadas.append(reglasPPT(0,-1,1))  # Piedra
jugadas.append(reglasPPT(1,0,-1))  # Papel
jugadas.append(reglasPPT(-1,1,0))  # Tijera
```

```python
for i in range(10):
  juego1 = np.random.randint(0,3)
  juego2 = np.random.randint(0,3)
  print("-", "Juego",i, "[ ",personajes[juego1],
        "VS", personajes[juego2],"]---------------")
  resultados = []
  resultados.append(jugadas[juego1].regresaLista())
  cadenaResultado = resultado[resultados[(0)][juego2]]
  print("          <",personajes[juego1],
        cadenaResultado, personajes[juego2], ">")
```

```python
def nombreJugador(id):
  return personajes[id]

def cadenaImagen(nombreArchivo):
  return "/content/"+nombreArchivo+".png"

```

```python
imgJugador1 = cv2.imread("/content/papel.png")
imgJugador2 = cv2.imread("/content/piedra.png")
imgEspacio  = cv2.imread("/content/espacio.png")
imgGanador  = cv2.imread("/content/gana.png")

imgConcat = cv2.hconcat([imgJugador1, imgJugador2, imgEspacio, imgGanador])
plt.imshow(imgConcat)
plt.show()
```

```python
juego1 = np.random.randint(0,3)
juego2 = np.random.randint(0,3)
rutaJugador1 = cadenaImagen(nombreJugador(juego1))
rutaJugador2 = cadenaImagen(nombreJugador(juego2))
print(rutaJugador1)
print(rutaJugador2)
imgJugador1 = cv2.imread(rutaJugador1)
imgJugador2 = cv2.imread(rutaJugador2)
imgEspacio  = cv2.imread("/content/espacio.png")

resultados = []
resultados.append(jugadas[juego1].regresaLista())
cadenaResultado = resultado[resultados[(0)][juego2]]

imgGanador  = cv2.imread(cadenaImagen(cadenaResultado))

imgConcat = cv2.hconcat([imgJugador1, imgJugador2, imgEspacio, imgGanador])
plt.imshow(imgConcat)
plt.show()
```
